```python
df = pd.read_csv('Lost__found__adoptable_pets.csv')
data = df[['Current_Location','animal_type','Animal_Gender','Date','City','State','Zip', 'Image','Memo']]
data
```

```python
data = data.dropna(subset=['City'])
data
```

```python
data = data.dropna(subset=['State'])
data
```

```python
data = data.dropna(subset=['Zip'])
data
```

```python
def categorize_animals_type(animal):
    if 'cat' in animal.lower():
        return 'Cat'
    elif 'dog' in animal.lower():
        return 'Dog'
    else:
        return 'Others'
```

```python
def categorize_animals_gender(animal):
   
    if pd.notna(animal) and 'Female' in animal:
        return 'Female'
    elif pd.notna(animal) and 'Male' in animal:
        return 'Male'
    else:
        return 'Male'
    
```

```python
# Apply the custom function to the 'Animal' column and create a new 'Category' column
data['animal_type'] = data['animal_type'].apply(categorize_animals_type)
data
```

```python
data['Animal_Gender'] = data['Animal_Gender'].apply(categorize_animals_gender)
data
```

```python
data['Address'] = data['City'] + ',' + data['State'] + ',' + data['Zip']
data
```
